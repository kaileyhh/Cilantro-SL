### MAPPING from ensembl ID's to genept embeddings in `GenePT_emb_ensembl.pkl`

```python
import pickle
from tqdm import tqdm
import pandas as pd
import torch
import pandas as pd
from sklearn.decomposition import PCA
```

```python
with open(f"/work/magroup/kaileyhu/Geneformer/geneformer/token_dictionary_gc95M.pkl", 'rb') as handle:
    token_dictionary = pickle.load(handle)

with open(f"/work/magroup/kaileyhu/Geneformer/geneformer/gene_name_id_dict_gc95M.pkl", 'rb') as handle:
    ensembl_dict = pickle.load(handle)
```

```python
genePT_file = "GenePT_gene_embedding_ada_text.pickle"
file_path = "/work/magroup/kaileyhu/data/genePT/GenePT_emebdding_v2"
```

```python
with open(f"{file_path}/{genePT_file}", 'rb') as handle:
    genept_embs = pickle.load(handle)
```

```python
ensembl_to_genept = {}
missing = []
for gene in tqdm(genept_embs):
    if gene not in ensembl_dict:
        missing.append(gene)
    else:
        ensembl_to_genept[ensembl_dict[gene]] = genept_embs[gene]
```

```python
print(f"{len(missing)} genes not in geneformer mapping dictionary")
```

```python
with open(f"{file_path}/GenePT_ada_ensembl.pkl", 'wb') as handle:
    pickle.dump(ensembl_to_genept, handle)
```

```python
genePT_file = "GenePT_gene_protein_embedding_model_3_text.pickle"

with open(f"{file_path}/{genePT_file}", 'rb') as handle:
    genept_embs = pickle.load(handle)
```

### PCA dimensionality reduction

```python
genept_df = pd.DataFrame(ensembl_to_genept).transpose()
```

```python
pca = PCA(n_components=128)

principalComponents = pca.fit_transform(genept_df)
```

```python
pca_df = pd.DataFrame(data = principalComponents, columns = [f"pc {i}" for i in range(128)])
```

```python
pca_df.index = genept_df.index
```

```python
pca_df.to_csv(f"{file_path}/genePT_ada_pca_128.csv")
```
