```python
import pandas as pd
import pickle
```

```python
df = pd.read_csv("/work/magroup/kaileyhu/SynLethDB/gene_sl_gene.tsv", sep='\t')
```

```python
ensembl_path = "/work/magroup/kaileyhu/Geneformer/geneformer/ensembl_mapping_dict_gc95M.pkl"

def invert_dict(dict_obj):
    return {v: k for k, v in dict_obj.items()}

with open(ensembl_path, "rb") as f:
    id_gene_dict = pickle.load(f)
    gene_id_dict = invert_dict(id_gene_dict)
```

```python
df["x_name"] = df["x_name"].replace(id_gene_dict)
```

```python
df["y_name"] = df["y_name"].replace(id_gene_dict)
```

```python
df.to_csv("/work/magroup/kaileyhu/SynLethDB/SL_with_ensembl.tsv")
```
