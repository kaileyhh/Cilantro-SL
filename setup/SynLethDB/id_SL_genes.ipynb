```python
ensembl_path = "/work/magroup/kaileyhu/Geneformer/geneformer/ensembl_mapping_dict_gc95M.pkl"

def inv_dict(dict_obj):
    res = {}
    counter = 0
    for k in dict_obj:
        v = dict_obj[k]
        if v in res:
            if v in res:
                continue
        res[v] = k

    return res

with open(ensembl_path, "rb") as f:
    id_gene_dict = pickle.load(f)

gene_id_dict = inv_dict(id_gene_dict)
```

```python
relevant_genes = ([df["x_name"].tolist()] + [df["y_name"].tolist()])[0]
```

```python
unique_genes = list(set(relevant_genes))
```

```python
f = open("/work/magroup/kaileyhu/SynLethDB/relevant_SL_genes.pkl","wb")
pickle.dump(unique_genes, f)
f.close()
```
