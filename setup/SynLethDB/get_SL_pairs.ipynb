```python
import pandas as pd
import pickle as pkl
```

```python
SL_df = pd.read_csv("/work/magroup/kaileyhu/datasets/SynLethDB/Human_SL.csv")
```

```python
set(SL_df['r.source'])
```

```python
top_fifth = SL_df.sort_values(by=['r.statistic_score'], ascending=False)
```

```python
top_fifth = top_fifth[: int(len(top_fifth)/5 + 100)]
```

```python
top_fifth.iloc[len(top_fifth)-2:]
```

```python
res_pairs = []
for i in range(len(top_fifth)):
    row = SL_df.iloc[i]
    g1 = row["n1.name"]
    g2 = row["n2.name"]
    res_pairs.append((g1, g2))
```

```python
with open ("/work/magroup/kaileyhu/datasets/SynLethDB/human_SL_top_fifth.pkl", "wb") as f:
    pkl.dump(res_pairs, f)
```

```python
with open ("/work/magroup/kaileyhu/datasets/SynLethDB/human_SL_brca_parp.pkl", "wb") as f:
    pkl.dump([("BRCA1", "PARP1"), ("BRCA1", "PARP2"), ("BRCA2", "PARP1"), ("BRCA2", "PARP2")], f)
```

```python
ensembl_path = "/work/magroup/kaileyhu/Geneformer/geneformer/ensembl_mapping_dict_gc95M.pkl"

def invert_dict(dict_obj):
    return {v: k for k, v in dict_obj.items()}

with open(ensembl_path, "rb") as f:
    id_gene_dict = pkl.load(f)
```

```python
res_pairs_ensembl = []
to_save = [("BRCA1", "PARP1"), ("BRCA1", "PARP2"), ("BRCA2", "PARP1"), ("BRCA2", "PARP2")]

for i in to_save:
    (g1, g2) = i
    g1 = id_gene_dict[g1]
    g2 = id_gene_dict[g2]
    res_pairs_ensembl.append((g1, g2))
```

```python
with open ("/work/magroup/kaileyhu/datasets/SynLethDB/human_SL_brca_parp_ENSEMBL.pkl", "wb") as f:
    pkl.dump(res_pairs_ensembl, f)
```

```python
res_pairs_ensembl = []
for i in range(len(top_fifth)):
    row = SL_df.iloc[i]
    try: 
        g1 = id_gene_dict[row["n1.name"]]
        g2 = id_gene_dict[row["n2.name"]]
        res_pairs_ensembl.append((g1, g2))
    except:
        print(f"Pair {row} not found")
```

```python
with open ("/work/magroup/kaileyhu/datasets/SynLethDB/human_SL_top_fifth_ENSEMBL.pkl", "wb") as f:
    pkl.dump(res_pairs_ensembl, f)
```

```python
all_genes = []
for i in range(len(SL_df)):
    row = SL_df.iloc[i]
    try: 
        g1 = id_gene_dict[row["n1.name"]]
        g2 = id_gene_dict[row["n2.name"]]
        all_genes.append(g1)
        all_genes.append(g2)
    except:
        continue

all_genes = list(set(all_genes))
```

```python
with open ("/work/magroup/kaileyhu/res/viable_files/Human_SL_genes.pkl", "wb") as f:
    pkl.dump(all_genes, f)
```

```python
pair_dict = {}
for i in range(len(SL_df)):
    row = SL_df.iloc[i]
    g1 = row["n1.name"]
    g2 = row["n2.name"]
    if g1 not in pair_dict:
        pair_dict[g1] = [g2]
    else:
        if g2 not in pair_dict[g1]:
            pair_dict[g1].append(g2)

    if g2 not in pair_dict:
        pair_dict[g2] = [g1]
    else:
        if g1 not in pair_dict[g2]:
            pair_dict[g2].append(g1)
```
