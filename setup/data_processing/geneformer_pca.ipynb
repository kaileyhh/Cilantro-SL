## Dimensionality reduction for the geneformer embeddings of DepMap data

```python
import pandas as pd
from sklearn.decomposition import PCA
```

```python
path = "/work/magroup/kaileyhu/res/perturbed/gf_12L_30M_i2048_SL/generated_df"
delta_embs = pd.read_hdf(f"{path}/delta_embs.h5", "table")
```

```python
dims = delta_embs.iloc[:, :512]
```

```python
pca = PCA(n_components=32)

principalComponents = pca.fit_transform(dims)
```

```python
pca_df = pd.DataFrame(data = principalComponents, columns = [f"{i}" for i in range(32)])
```

```python
pca_df['viability score'] = delta_embs['viability score']
```

```python
pca_df.index = delta_embs.index
```

```python
pca_df.to_hdf(f"{path}/delta_embs_pca_32.h5", "table")
```
