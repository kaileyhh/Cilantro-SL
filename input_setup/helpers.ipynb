```python
import numpy as np
import pandas as pd

import sys
import os 
import pickle as pkl

sys.path.append('/work/magroup/kaileyhu/Geneformer')

from geneformer.tokenizer import TOKEN_DICTIONARY_FILE
from datasets import load_from_disk, concatenate_datasets
```

```python
### Rename columns of dataframe to Geneformer tokens

def rename_cols (df):
    ensembl_df = pd.read_csv('/work/magroup/ehaber/geneformer_experiments/gene_info_table.csv')
    
    col = ['gene_name', 'ensembl_id']
    ensembl_df = ensembl_df[col]
    value_mapping = dict(zip(ensembl_df['gene_name'], ensembl_df['ensembl_id']))
    
    ds = load_from_disk("/work/magroup/kaileyhu/res/hvg_500_tokenized.dataset")
    
    with open(TOKEN_DICTIONARY_FILE, "rb") as td:
        gene_token_dict = pkl.load(td)
    
    gene_to_token = {}
    
    for col in df.columns:
        if col.split(" ")[0] in value_mapping:
            if value_mapping[col.split(" ")[0]] in gene_token_dict:
                gene_to_token[col] = gene_token_dict[value_mapping[col.split(" ")[0]]]

    return gene_to_token
```
