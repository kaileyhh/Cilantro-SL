```python
import sys
import os

from tqdm import tqdm
import pandas as pd
import numpy as np
import pickle as pkl
```

```python
with open ("/work/magroup/kaileyhu/res/perturbed/gf_12L_30M_i2048_SL/generated_df/sub_embs.pkl", "rb") as f:
    a = pkl.load(f)
```

```python
full_dict = {}
for gene in tqdm(a):
    df_temp = a[gene]
    def fix_name(x):
        return f"({x}, {gene})"
    df_temp.index = list(map(fix_name, df_temp.index))
    d = df_temp.to_dict('index')
    full_dict |= d
```

```python
df = pd.DataFrame(full_dict)
```

```python
df = df.T
```
