### Construct a dataframe of the following form:
- Rows: `(patient, gene)` pairs
- Columns: 
    - First 512 columns: 512-dimension embedding space from geneformer
    - Last column: associated viability score with knocking out `gene` from `patient`

Using the data saved in the dictionaries generated by `./ko_to_adata.ipynb`

```python
import pandas as pd 
import numpy as np
import anndata as ad
from tqdm import tqdm
import pickle
```

```python
emb_file = "/work/magroup/kaileyhu/res/via_classifier/emb_diffs.pkl"
via_file = "/work/magroup/kaileyhu/res/via_classifier/viability_dict.pkl"
```

```python
with open(emb_file, 'rb') as f:
    sub_embs = pickle.load(f)

with open(via_file, 'rb') as f:
    viability = pickle.load(f)
```

```python
diff_dict = {}

for gene in tqdm(sub_embs):
    df_diff = sub_embs[gene]
    for patient in df_diff.index:
        if (patient, gene) in viability:
            temp = (df_diff.loc[patient].tolist())
            temp.append(viability[(patient, gene)])
            diff_dict[(patient, gene)] = temp
```

```python
df = pd.DataFrame.from_dict(diff_dict, orient='index')
```

```python
df.rename(columns = {512 : "viability score"}, inplace = True)
```
