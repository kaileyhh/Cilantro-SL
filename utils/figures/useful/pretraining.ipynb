```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle as pkl
import numpy as np
```

```python
name = "pretraining_stats"
cv = 1
metrics_dict_path = f"/work/magroup/kaileyhu/res/ablations/cv{cv}/{name}.pkl"

with open (metrics_dict_path, "rb") as f:
    metrics = pkl.load(f)
```

```python
metrics = metrics[0]
```

```python
y_pred, y_true = metrics
```

```python
plt.scatter(y_true, y_pred, s=1, color='#2A9D8F')
plt.xlabel("True viability score")
plt.ylabel("Predicted viability score")

min_val = min(np.min(y_true), np.min(y_pred))
max_val = max(np.max(y_true), np.max(y_pred))

line_range = [min_val - 0.5, max_val + 0.5]

plt.plot(line_range, line_range, color='#E76F51', linestyle='--', linewidth=2, label='Perfect prediction')

plt.title("Viability pretraining performance")
```
