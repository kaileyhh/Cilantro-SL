```python
def compute_stds(name):
    with open (f"{pfx3}{name}", "rb") as f:
        temp = pkl.load(f)
    indiv_results = temp['individual_results']
    res = {}
    for k in indiv_results:
        res[k] = np.std(indiv_results[k])
    print(f"aupr: {np.mean(indiv_results['f1'])}")
    print(f"{pfx3} {name}: {res['f1']}")
    print(f"{name} illustrator: {ratio * res['f1']}")
```

```python
import pandas as pd
from scipy import stats

data = {
    'Fold': [1, 2, 3, 4, 5],
    'Delta & gene': [0.88804391598719, 0.9409459146593993, 0.9372402686321963, 0.8934369690594616, 0.8985630552156888],
    'Perturbed & gene': [0.8810657087584198, 0.9048643739598519, 0.9103844666463299, 0.8868764963284735, 0.8734524838833629],
    'Delta only': [0.8370331177204877, 0.9041037579722152, 0.8446530109108932, 0.8831766044733522, 0.907634913031941]
}
df = pd.DataFrame(data)

# --- 2. Compare "Delta & gene" vs. "Perturbed & gene" ---
method_a = df['Delta & gene']
method_b = df['Perturbed & gene']

# Run the paired t-test
t_statistic_1, p_value_1 = stats.ttest_rel(method_a, method_b)

print(f"\n--- Comparison 1: Delta & gene vs. Perturbed & gene ---")
print(f"Paired T-Statistic: {t_statistic_1:.4f}")
print(f"P-Value: {p_value_1:.4f}")
print("-" * 40)

method_c = df['Delta only']
t_statistic_2, p_value_2 = stats.ttest_rel(method_a, method_c)

print(f"\n--- Comparison 2: Delta & gene vs. Delta only ---")
print(f"Paired T-Statistic: {t_statistic_2:.4f}")
print(f"P-Value: {p_value_2:.4f}")
print("-" * 40)

t_statistic_3, p_value_3 = stats.ttest_rel(method_b, method_c)

print(f"\n--- Comparison 3: Perturbed & gene vs. Delta only ---")
print(f"Paired T-Statistic: {t_statistic_3:.4f}")
print(f"P-Value: {p_value_3:.4f}")
print("-" * 40)
```
