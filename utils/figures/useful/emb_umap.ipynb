```python
df2_embs = df2.drop(columns = ['patient', 'gene 1', 'gene 2', 'gene 1 ensembl', 'gene 2 ensembl', 'SL'])
```

```python
adata = sc.AnnData(df2_embs)
```

```python
adata.obs['label'] = list(map(lambda x : 'SL' if x else 'non SL', list(df2['SL'])))
```
